# Import Stuff

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(font_scale=1.5)
```

```python
vacuum_cp = np.array([0.0, 0.06, 0.33, 0.42, 0.52]) # From vacuum/vacuum_comparison.ipynb
beryl_cp = np.array([0.0, 0.07, 0.27, 0.43, 0.52]) # From fine_sweep/analyze_fine_sweep.ipynp
```

# Compare

```python
labels = 'A B C D E'.split()
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.gca()
data = 1*(vacuum_cp-beryl_cp)
sns.barplot(x=labels,y=data,ax=ax)
sns.lineplot(x=labels,y=data,ax=ax)
ax.set(xlabel='Critical Point',ylabel='Angle difference (Vaccum$-$Beryl) [rad]')
```
