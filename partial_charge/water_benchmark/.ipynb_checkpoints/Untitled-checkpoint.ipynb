# Setup

## Import Stuff

```python
import pandas as pd 
import numpy as np 
import pickle as pkl
import seaborn as sns 
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

with open("tidy_data/pc_water_benchmark.pkl","rb") as _filein:
    raw_data = pkl.load(_filein)

raw_df = pd.DataFrame(raw_data)
raw_df.head()
```

## Anonymous Functions

```python
get_sys_info = lambda param,df=raw_df : [i[param] for i in df.sys_info.values]
filter_by_system = lambda sys,df=raw_df: df[df.system == sys]
filter_by_kpoints = lambda kpts,df: df[df.kpts == kpts].sort_values('run',axis=0)
```

## Some Data Cleaning

```python
raw_df['kpts'] = get_sys_info("kpts")
raw_df['system'] = get_sys_info("sys")
raw_df['run'] = get_sys_info("run")
raw_df.kpts = raw_df.kpts.apply(int)
raw_df.run = raw_df.run.apply(int)
raw_df.drop('sys_info',axis=1,inplace=True)
```

```python
filter_by_kpoints(5,filter_by_system("monomer"))
```
